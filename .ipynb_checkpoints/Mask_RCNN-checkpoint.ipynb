```python
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
import os
print("done")
```

```python

# --- Load class names ---
with open("mask_rcnn_classes.txt", "r") as f:
    classes = f.read().strip().split("\n")

# --- Load pre-trained Mask R-CNN model ---
net = cv2.dnn.readNetFromTensorflow(
    "dnn/frozen_inference_graph.pb",
    "dnn/mask_rcnn_inception_v2_coco_2018_01_28.pbtxt"
)

# --- Load image ---
img = cv2.imread("car.jpg")
height, width, _ = img.shape

# --- Copies for visualization ---
img_boxes = img.copy()
img_masks = img.copy()

# --- Generate unique random colors for all 90 COCO classes ---
colors = [tuple(np.random.randint(0, 256, 3).tolist()) for _ in range(90)]

# --- Create blob and run the network ---
blob = cv2.dnn.blobFromImage(img, swapRB=True, crop=False)
net.setInput(blob)
boxes, masks = net.forward(["detection_out_final", "detection_masks"])

# --- Evaluation metrics ---
def compute_iou(pred_mask, gt_mask):
    intersection = np.logical_and(pred_mask, gt_mask).sum()
    union = np.logical_or(pred_mask, gt_mask).sum()
    return intersection / union if union != 0 else 0

def compute_precision_recall(pred_mask, gt_mask):
    TP = np.logical_and(pred_mask == 1, gt_mask == 1).sum()
    FP = np.logical_and(pred_mask == 1, gt_mask == 0).sum()
    FN = np.logical_and(pred_mask == 0, gt_mask == 1).sum()

    precision = TP / (TP + FP) if (TP + FP) else 0
    recall = TP / (TP + FN) if (TP + FN) else 0

    return precision, recall

# --- Create ground_truth directory ---
os.makedirs("ground_truth", exist_ok=True)

# --- Process detections ---
num_detections = boxes.shape[2]
ious, precisions, recalls = [], [], []

for i in range(num_detections):
    box = boxes[0, 0, i]
    score = box[2]

    if score > 0.2:
        class_id = int(box[1])
        class_name = classes[class_id] if class_id < len(classes) else "Unknown"
        color = colors[class_id % len(colors)]

        # Get bounding box
        x1 = int(box[3] * width)
        y1 = int(box[4] * height)
        x2 = int(box[5] * width)
        y2 = int(box[6] * height)

        # Add padding
        padding = 10
        x1 = max(0, x1 - padding)
        y1 = max(0, y1 - padding)
        x2 = min(width, x2 + padding)
        y2 = min(height, y2 + padding)

        # Draw bounding box
        cv2.rectangle(img_boxes, (x1, y1), (x2, y2), color, 2)
        label = f"{class_name}: {round(score * 100, 1)}%"
        cv2.putText(img_boxes, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

        # Resize predicted mask
        pred_mask = masks[i, class_id]
        pred_mask = cv2.resize(pred_mask, (x2 - x1, y2 - y1))
        pred_mask = (pred_mask > 0.5).astype(np.uint8)

        # Apply mask on the region
        roi = img_masks[y1:y2, x1:x2]
        color_mask = np.zeros_like(roi, dtype=np.uint8)
        for c in range(3):
            color_mask[:, :, c] = color[c]
        color_mask = color_mask * pred_mask[:, :, np.newaxis]
        blended = cv2.addWeighted(roi, 1.0, color_mask, 0.5, 0)
        img_masks[y1:y2, x1:x2] = np.where(pred_mask[:, :, np.newaxis], blended, roi)

        # Save mask as mock ground truth
        mask_path = f"ground_truth/mask_{i+1}.png"
        cv2.imwrite(mask_path, (pred_mask * 255).astype(np.uint8))

        # Evaluate predicted mask against saved version (should be perfect)
        gt_mask = pred_mask.copy()
        iou = compute_iou(pred_mask, gt_mask)
        precision, recall = compute_precision_recall(pred_mask, gt_mask)

        ious.append(iou)
        precisions.append(precision)
        recalls.append(recall)

        print(f"[{class_name}] Detection {i+1} — IoU: {iou:.3f}, Precision: {precision:.3f}, Recall: {recall:.3f}")

# --- Evaluation summary ---
if ious:
    print("\n=== Evaluation Summary ===")
    print(f"Mean IoU: {np.mean(ious):.3f}")
    print(f"Mean Precision: {np.mean(precisions):.3f}")
    print(f"Mean Recall: {np.mean(recalls):.3f}")
else:
    print("No valid detections found.")

# --- Convert BGR to RGB for plotting ---
img_boxes_rgb = cv2.cvtColor(img_boxes, cv2.COLOR_BGR2RGB)
img_masks_rgb = cv2.cvtColor(img_masks, cv2.COLOR_BGR2RGB)

# --- Show results ---
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_boxes_rgb)
plt.axis("off")
plt.title("Bounding Boxes")

plt.subplot(1, 2, 2)
plt.imshow(img_masks_rgb)
plt.axis("off")
plt.title("Segmentation Masks")

plt.tight_layout()
plt.show()
```
